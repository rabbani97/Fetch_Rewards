Receipts.CSV


```python
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/Fetch/receipts.csv'
data = pd.read_csv(file_path)
```

```python
# Standardize column names
data.columns = data.columns.str.lower().str.replace(' ', '_')
```

```python
# Handle missing values
# Fill numerical columns with the median
numerical_cols = data.select_dtypes(include=['float64']).columns
for col in numerical_cols:
    data[col].fillna(data[col].median(), inplace=True)
```

```python
# Fill categorical columns with 'Unknown'
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    data[col].fillna('Unknown', inplace=True)
```

```python
# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Convert date columns to datetime
date_cols = ['date_scanned', 'finished_date', 'modify_date', 'points_awarded_date', 'purchase_date']
for col in date_cols:
    data[col] = pd.to_datetime(data[col], errors='coerce')
```

```python
# Detect and handle outliers
# For this example, we'll cap outliers at the 95th percentile
def cap_outliers(series):
    upper_limit = series.quantile(0.95)
    return series.apply(lambda x: min(x, upper_limit))

for col in numerical_cols:
    data[col] = cap_outliers(data[col])
```

```python
# Correct data types for numerical columns
data['purchaseditemcount'] = data['purchaseditemcount'].astype('int', errors='ignore')
data['totalspent'] = data['totalspent'].astype('float', errors='ignore')
```

```python
# Validate and correct data
# Example: Ensuring 'rewardsreceiptstatus' only contains valid statuses
valid_statuses = ['FINISHED', 'REJECTED', 'PENDING']
data = data[data['rewardsreceiptstatus'].isin(valid_statuses)]
```

```python
# Display cleaned data information
print(data.info())
print(data.head())
```

```python
# Save the cleaned data to a new CSV file
cleaned_file_path = '/content/drive/MyDrive/Fetch/cleaned_receipts.csv'
data.to_csv(cleaned_file_path, index=False)
```

```python
# Detect and handle outliers
# Cap outliers at the 95th percentile
def cap_outliers(series):
    upper_limit = series.quantile(0.95)
    return series.apply(lambda x: min(x, upper_limit))

for col in numerical_cols:
    data[col] = cap_outliers(data[col])
```

```python
# Simulate Users Data
users_data = {
    'user_id': [i for i in range(1, 101)],
    'creation_date': [datetime.today() - timedelta(days=(i % 180)) for i in range(1, 101)]
}

users_df = pd.DataFrame(users_data)
```

```python
# Merge brand data to transactions
transactions_df = transactions_df.merge(brands_df, left_on='brand_id', right_index=True)
```

```python
# Convert dates to ensure consistent comparison
receipts_df['date'] = pd.to_datetime(receipts_df['date'])
users_df['creation_date'] = pd.to_datetime(users_df['creation_date'])
```

```python
# Helper function to get the start and end of the most recent month and the previous month
def get_month_ranges():
    today = datetime.today()
    current_month_start = today.replace(day=1)
    previous_month_end = current_month_start - timedelta(days=1)
    previous_month_start = previous_month_end.replace(day=1)

    return current_month_start, previous_month_start, previous_month_end
```

```python
# Get the month ranges
current_month_start, previous_month_start, previous_month_end = get_month_ranges()
```

```python
# Convert to strings for filtering
current_month_start_str = current_month_start.strftime('%Y-%m-%d')
previous_month_start_str = previous_month_start.strftime('%Y-%m-%d')
previous_month_end_str = previous_month_end.strftime('%Y-%m-%d')
```

## Question 1:
# Top 5 brands by receipts scanned for most recent month




```python
# Answer the questions

recent_month_receipts = receipts_df[receipts_df['date'] >= current_month_start]
top_brands_recent_month = transactions_df[transactions_df['receipt_id'].isin(recent_month_receipts['receipt_id'])] \
    .groupby('NAME')['transaction_id'].count() \
    .sort_values(ascending=False).head(5)
print("Top 5 brands by receipts scanned for most recent month:")
print(top_brands_recent_month)
```

# Question 2:
`How does the ranking of the top 5 brands by receipts scanned for the recent month compare to the ranking for the previous month?`

```python
# Question 2: Compare to previous month
previous_month_receipts = receipts_df[(receipts_df['date'] >= previous_month_start) & (receipts_df['date'] <= previous_month_end)]
top_brands_previous_month = transactions_df[transactions_df['receipt_id'].isin(previous_month_receipts['receipt_id'])] \
    .groupby('NAME')['transaction_id'].count() \
    .sort_values(ascending=False).head(5)
print("\nTop 5 brands by receipts scanned for the previous month:")
print(top_brands_previous_month)
```

# `Question 3:`
`When considering average spend from receipts with 'rewardsReceiptStatus’ of ‘Accepted’ or ‘Rejected’, which is greater?`

```python

accepted_rejected_receipts = receipts_df[receipts_df['status'].isin(['Accepted', 'Rejected'])]
avg_spend_by_status = accepted_rejected_receipts.groupby('status')['total_spend'].mean()
print("\nAverage spend by receipt status:")
print(avg_spend_by_status)
```

# `Question 4:`
`When considering total number of items purchased from receipts with 'rewardsReceiptStatus’ of ‘Accepted’ or ‘Rejected’, which is greater?`

```python

items_by_status = transactions_df[transactions_df['receipt_id'].isin(accepted_rejected_receipts['receipt_id'])] \
    .merge(accepted_rejected_receipts[['receipt_id', 'status']], on='receipt_id') \
    .groupby('status')['item_count'].sum()
print("\nTotal items purchased by receipt status:")
print(items_by_status)
```

# `Question 5:`
`Which brand has the most spend among users who were created within the past 6 months?`




```python

six_months_ago = datetime.today() - timedelta(days=180)
recent_users = users_df[users_df['creation_date'] >= six_months_ago]['user_id']
recent_user_receipts = receipts_df[receipts_df['user_id'].isin(recent_users)]
most_spend_brand_recent_users = transactions_df[transactions_df['receipt_id'].isin(recent_user_receipts['receipt_id'])] \
    .groupby('NAME')['spend'].sum() \
    .sort_values(ascending=False).head(1)
print("\nBrand with the most spend among recent users:")
print(most_spend_brand_recent_users)
```

# `Question 6:`

`Which brand has the most transactions among users who were created within the past 6 months?`

```python
# Question 6: Brand with the most transactions among recent users
most_transactions_brand_recent_users = transactions_df[transactions_df['receipt_id'].isin(recent_user_receipts['receipt_id'])] \
    .groupby('NAME')['transaction_id'].count() \
    .sort_values(ascending=False).head(1)
print("\nBrand with the most transactions among recent users:")
print(most_transactions_brand_recent_users)
```
